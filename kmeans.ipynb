```python
import torch
import sklearn
from sklearn import datasets, cluster
import matplotlib.pyplot as plt

from KMeans import KMeans
```

```python
X, y = sklearn.datasets.make_blobs(n_samples=10000, n_features=2, centers=20, cluster_std=0.4)
X = torch.tensor(X)
```

My KMeans

```python
device = 'cuda' if torch.cuda.is_available() else 'cpu'
```

```python
centroids1 = (KMeans(20).fit_predict(X.to(device).unsqueeze(0)).cpu().numpy()).squeeze(0)
```

```python
plt.scatter(X[:, 0], X[:, 1], c=y, s=0.3, cmap='jet')
plt.scatter(centroids1[:, 0], centroids1[:, 1], s=30, c='black')
plt.show()
```

Sklearn KMeans

```python
centroids2 = (sklearn.cluster.KMeans(20).fit(X).cluster_centers_)
```

```python
plt.scatter(X[:, 0], X[:, 1], c=y, s=0.3, cmap='jet')
plt.scatter(centroids2[:, 0], centroids2[:, 1], s=30, c='black')
plt.show()
```
